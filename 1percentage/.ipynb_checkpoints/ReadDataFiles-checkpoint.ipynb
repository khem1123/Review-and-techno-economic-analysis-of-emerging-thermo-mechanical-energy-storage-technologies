```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('LIquidAirRegu.csv')
df.head(10)
```

```python
a=np.zeros(len(df.storage))
for i in np.arange(len(df.storage)):
    a[i]=float(df.storage[i].strip("[]"))
df.storage=a
```

```python
a=np.zeros(len(df.profit))
for i in np.arange(len(df.profit)):
    a[i]=float(df.profit[i].strip("[]"))
df.profit=a
```

```python
A=df.query('profit<100 & profit>0 ').reset_index()
A=A.drop_duplicates(subset=['energy','effi'], keep='last').reset_index()
import openpyxl
from openpyxl import load_workbook

A.to_csv('PTPReg')  
  
```
